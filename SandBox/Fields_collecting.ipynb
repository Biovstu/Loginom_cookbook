```python
refs = 'Ref_Key,Description,ДатаНачала,Партнер_Key,Партнер____Presentation,био_Заказчик_Key,био_Заказчик____Presentation,био_ОкончаниеСрокаПодачиЗаявок,био_ОтветственныйЗаТендер_Key,био_ОтветственныйЗаТендер____Presentation,Комментарий,Статус,био_СтатусТендера,DeletionMark,био_ТипСделки,био_ОтветМенеджера,Code,био_НачальнаяЦена,био_КонечнаяЦена,био_Город,био_ДатаОтправкиЗапросаЗаказчику,био_ЗапросНаРазъясненияВыполнялся,био_ЗапросОтправленЗаказчику,био_ИтогиКонкурса,био_ИтогиПодачиЖалобы,био_КазначейскоеИлиБанковскоеСопровождение,био_КоличествоКонкурентов,био_КоличествоТендеров,био_КомментарийПоОтвету,био_КонкурентовНет,био_КонкурентыВнесены,био_КонтрактПоЖизнеобеспечению,био_МенеджерУстановившийОтвет_Key,био_МенеджерУстановившийОтвет____Presentation,био_Менеджеры,био_Моноаукцион,био_НомерИДатаДоговора,био_ОбеспечениеГарантииНаОборудование,био_ПлановаяДатаТоргов,био_Приказ102,био_Приказ155,био_Приказ341,био_Приказ616,био_Приказ617,био_ПродуктовыеНаправленияСтрокой,био_РазницаКонечнойИНачальнойЦены,био_СМП,био_СоответствиеПоТЗ,био_СпособТоргов_Key,био_СпособТоргов____Presentation,био_СсылкаНаТендернуюПлощадку,био_СтадияКонтракта,био_Счет_Key,био_СчетДилера_Key,био_СчетДилера____Presentation,био_ТекущееРешениеПоУчастию,био_ТендерКод,био_ТЗ_С_Привязкой_К_КТРУ,био_ФактическаяДатаКонкурса,био_ФактическаяДатаПодачиЗаявки,био_ФедеральныйОкруг_Key,био_ФедеральныйОкруг____Presentation,ВидСделки_Key,ВидСделки____Presentation,ДатаОкончания,ДатаРегистрации'
refs = refs.replace('Ref_Key,','').replace('Ref_Key','')
ref = refs.split(',')
dif = 15
ref_dif = []
if len(ref) // dif > 0:
    for i in range(len(ref) // dif):
        ref_dif.append(','.join(ref[i * dif:i * dif+dif]))
if len(ref) % dif != 0:
    ref_dif.append(','.join(ref[len(ref)- (len(ref) % dif):]))
download_data = pd.DataFrame({})
for i,col in enumerate(ref_dif):
    if col != 'Ref_Key':
        buff_resp = requests.get(f"http://dcr-iis02/erp_data/odata/standard.odata/Catalog_БИТ_CRM_3_СделкиСКлиентами?$select=Ref_Key,{col}", headers=dict(Accept='application/json;odata=nometadata'), auth=basic)
        buff_df = pd.read_json(json.dumps(buff_resp.json()['value']))
        if i == 0:
            download_data = buff_df
        else:
            download_data = download_data.merge(buff_df, left_on='Ref_Key', right_on='Ref_Key', how='left')
download_data
```
