```python
file_path = 'C:\\Users\\konstantin_kutovoy\\Desktop\\'
file_name_source = 'tf3.csv'
file_name_destination = 'tf3_cleared.csv'
source = open(file_path+file_name_source, mode = 'tr', encoding = 'utf-8')
destination = open(file_path + file_name_destination, mode = 'ta+', encoding = 'utf-8')
val = ''
for num, line in enumerate(source):
    if num == 0:
        qty = line.count(',')
    if qty <= line.count(','):
        tmp = line.replace('\f','').replace('\n','').replace('\r','').replace('\t','').replace('\v','')
        destination.write(f'{tmp}\n')
    else:
        val += line.replace('\f','').replace('\n','').replace('\r','').replace('\t','').replace('\v','')
        if qty <= val.count(','):
            destination.write(f"{val}\n")
            val = ''
source.close()
destination.close()
```

```python
file_name_destination_good = 'tf3_good.csv'
file_name_destination_bad = 'tf3_bad.csv'
source = open(file_path+file_name_destination, mode = 'tr', encoding = 'utf-8')
destination_good = open(file_path + file_name_destination_good, mode = 'ta+', encoding = 'utf-8')
destination_bad = open(file_path + file_name_destination_bad, mode = 'ta+', encoding = 'utf-8')
for num, line in enumerate(source):
    tmp = line.replace('\n','')
    if num == 0:
        destination_good.write(f'{tmp}\n')
        destination_bad.write(f'{tmp}\n')
    else:
        if line.split(',')[0].isdigit():
            destination_good.write(f'{tmp}\n')
        else:
            destination_bad.write(f'{tmp}\n')
source.close()
destination_good.close()
destination_bad.close()
```
