```python
import re
import pandas as pd
```

```python
string = 'У! ПЕТСАЛ®, Сыворотки диагностические сальмонеллезные адсорбированные для РА, лиофилизат для диагностических целей, моновалентная О- сыворотка  (7) №1'
symbols = re.findall(r'\W',string)
symbols
```

```python
d = {}
for c in symbols:
    lab = f'"{c}":char({ord(c)})'
    d.setdefault(lab,list(c))
    d[lab].append(c)
d
```

```python
dat = pd.DataFrame.from_dict(d, orient='index').T
dat
```

```python
pd.DataFrame(dat.nunique()).reset_index()
```
